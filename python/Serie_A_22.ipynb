```python
#import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the file
table = pd.read_csv('SerieA22.csv')
```

```python
#select the columns
my_table = table[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR']].copy()
```

```python
#obtain the full time stats
my_table['FTTG'] = my_table['FTHG'] + my_table['FTAG']
my_table['HTTG'] = my_table['HTHG'] + my_table['HTAG']
```

```python
#create the over 1.5 & under 3.5 dataframe
over_table = my_table[my_table['FTTG'] >= 2]
under_table = my_table[my_table['FTTG'] < 4]
```

```python
#calculate the number of matches played by the teams
a = my_table.HomeTeam.value_counts()
b = my_table.AwayTeam.value_counts()
```

```python
#create a df to display the teams and total matches played
matches_played = pd.DataFrame([a, b]).T
matches_played['Total'] = matches_played['HomeTeam'] + matches_played['AwayTeam']
```

```python
#obtain the count of the matches for each team they score over 1.5 goals
x = over_table['HomeTeam'].value_counts()
y = over_table['AwayTeam'].value_counts()
```

```python
#create a data frame for the these teams
matches_over = pd.DataFrame([x, y]).T
```

```python
#the count of matches for each team under 3.5 goals
x = under_table['HomeTeam'].value_counts()
y = under_table['AwayTeam'].value_counts()
```

```python
#create a data frame for the these teams
matches_under = pd.DataFrame([x, y]).T
```

```python
#create a totals table
matches_over['Total'] = matches_over['HomeTeam'] + matches_over['AwayTeam']
matches_under['Total'] = matches_under['HomeTeam'] + matches_under['AwayTeam']
matches_over['TMP'] = matches_played['Total']
matches_under['TMP'] = matches_played['Total']
```

```python
#calculate the percentage of the matches over 1.5 goals to all matches played
matches_over['Percentage'] = (matches_over['Total'] / matches_played['Total']) * 100
matches_under['Percentage'] = (matches_under['Total'] / matches_played['Total']) * 100
```

```python
#Home & Away percentage of over 1.5 goals
matches_over['HP'] = (matches_over['HomeTeam'] / matches_played['HomeTeam']) * 100
matches_over['AP'] = (matches_over['AwayTeam'] / matches_played['AwayTeam']) * 100

#Home & Away percentage of under 3.5 goals
matches_under['HP'] = (matches_under['HomeTeam'] / matches_played['HomeTeam']) * 100
matches_under['AP'] = (matches_under['AwayTeam'] / matches_played['AwayTeam']) * 100
```

```python
#create a function to print the home and away percentages for the matches that have over 1.5 goals
def team_percentages_over(home_p, away_p):
    print('These are the away team percentages for over 1.5 goals: \n')  
    for team, percentage in away_p.items():
        print(f'{team} : {percentage:.1f}')

    print('\nThese are the home team percentages for over 1.5 goals: \n')
    for team, percentage in home_p.items():
        print(f'{team} : {percentage:.1f}')

away_p = matches_over.AP
away_p = away_p.to_dict()
home_p = matches_over.HP
home_p = home_p.to_dict()

team_percentages_over(home_p, away_p)
```

```python
#create a function to print the home and away percentages for the matches that have under 3.5 goals
def team_percentages_under(home_p, away_p):
    print('These are the away team percentages for under 3.5 goals: \n')  
    for team, percentage in away_p.items():
        print(f'{team} : {percentage:.1f}')

    print('\nThese are the home team percentages for under 3.5 goals: \n')
    for team, percentage in home_p.items():
        print(f'{team} : {percentage:.1f}')

away_p = matches_under.AP
away_p = away_p.to_dict()
home_p = matches_under.HP
home_p = home_p.to_dict()

team_percentages_under(home_p, away_p)
```

```python
#display the overall dataframes
print('These is a table of the MATCHES with over 1.5 goals: \n')
print(matches_over)
print('\nThese is a table of the MATCHES with under 3.5 goals: \n')
print(matches_under)
```
